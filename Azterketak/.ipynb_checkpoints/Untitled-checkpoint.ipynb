## 1. Ariketa

Estatistikan, mediana hainbat datu txikienetik handienera ordenaturik daudela, erdian dagoen datua da. Datu-kopurua bakoitia bada, ondoko kasuan adibidez:

45,12,1,1,26

ordenatu ostean (1,1,12,26,45), datu hauetarako mediana 3. posiziokoa balioa da, 12 balioa hain zuzen ere. Datu-kopurua bikoitia denean ordea, adibidez:

45,12,1,1,26,35

ordenatu ostean (1,1,12,26,35,45) erdian bi datu egongo direnez, mediana definitzeko bi aukera ezberdin daude:

1.	Erdiko bi balioen batezbestekoa hartzea
2.	Erdiko bi balioetatik ausazko bat aukeratzea

Sortu ezazu `median(it, rand_if_odd=False)` funtzio bat, iteragarri baten elementuen mediana kalkulatuko duena. Iteragarriaren elementu kopurua bikoitia balitz, `rand_if_odd==False` denean erdiko bi elementuen batezbestekoa bueltatuko du eta `rand_if_odd==True` denean erdiko bi elementuetatik ausazko bat bueltatuko du. 

````python
from random import choice

def median(it, rand_if_odd=False):
    z = sorted(it)
    n = len(z)
    e = n//2
    if n%2 == 1:
        return z[e]
    elif rand_if_odd :
        return z[choice([e-1,e])]
    else :
        return (z[e-1]+z[e])/2 
````

Erabilpenaren adibide batzuk (hau ez da eskatzen):

````python
print(median([45,12,1,1,26]))
print(median([45,12,1,1,26,35]))
print(median([45,12,1,1,26,35], True))
print(median([45,12,1,1,26,35], True))
print(median([45,12,1,1,26,35], True))
print(median([45,12,1,1,26,35], True))
print(median([45,12,1,1,26,35], True))
print(median("eouai"))
print(median("eoai", True))
print(median("eoai", True))
print(median("eoai", True))
````

## 2. Ariketa

<u>**Aurreko funtzioa erabiliz**</u>, sortu ezazu funtzio bat, testu fitxategi baten bideizena eta kodifikazioa jasoko duena. Testu fitxategiko ilara bakoitzean pertsona baten izen-abizenak, adina, NAN zenbakia eta jaioterria egongo dira hutsunez bananduak:

```
    Leire Legarra Aizpurua 19 62342534G Zegama
    Aimar Lopez de Uribe Martinez de Gereño 21 73245384R Durango
    Izaro Leizaola Lopez 17 24394637J Gamiz-Fika
```

Funtzioak herri bakoitzeko adinen mediana adierazten duen hiztegia bueltatuko du (gakoa herri izena eta balioa herri horretako adinen mediana). Mediana kalkulatzean, adin kopurua bikoitia balitz, aurreko funtzioko 2. definizioa erabiliko da (erdiko balioetatik ausazko bat).

<u>**OHARRA-1:**</u> Izen abizenek eremu kopuru aldakorra izan dezakete, herriak ordea beti hitz bakar batez adieraziak egongo dira, hutsunerik gabe.

<u>**OHARRA-2:**</u> Ez da beharrezkoa aurreko funtzioa eginda izatea, existitzen dela suposa dezakezu.


````python
def adin_medianak(filename, encoding="utf8"):
    with open(filename, encoding=encoding) as f:
        d = {}
        for line in f:
            fields = line.split()
            age = fields[-3]
            city = fields[-1]
            d.setdefault(city,[]).append(age)
    for city,z in d.items():
        d[city] = median(z,True)
    return d
    #return {city:median(z,True) for city,z in d.items()}
````
