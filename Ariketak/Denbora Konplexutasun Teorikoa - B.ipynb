Zenbaki zerrenda batetako elementu bakoitien batura

````python
def bakoitien_batura(z):
    b = 0
    for x in z :
        if x%2 == 1 :
            b += x
    return b
````

Problemaren tamaina: $n=len(z)$

```python
def bakoitien_batura(z):
    b = 0              # 1
    for x in z :       # n x 1
        if x%2 == 1 :  # | n x 1
            b += x     # |
    return b           # 1
```

$t(n) = 2n + 2$ &rarr; orden zehatz lineala

````python
def bakoitien_batura(z):
    return sum(x for x in z if x%2 == 1 )
````

Problemaren tamaina: $n=len(z)$

* `sum(it)` &rarr; it-ren elementu kopuru pausu + 1
* `(x for x in ...)` generadorea:
   * sortzea &rarr; 1  
   * zeharkatzea &rarr; `next(g)` 
      * elementu guztiak bakoitiak &rarr; 1 pausu x n aldiz
      * elementu guztiak bikoitiak &rarr; n pausu x 1 aldiz
   
```python

```

Hiztegi edo multzorik erabili gabe, zerrenda batetan errepikatuak (bi edo gehiagotan) agertzen diren elementu ezberdinen kopurua.

````python
def errepikatu_kopurua(z):
    z2 = []
    z3 = []
    for x in z :
        if x not in z2 :
            z2.append(x)
        elif x not in z3 :
            z3.append(x)
    return len(z3)
````


* $t(n) = n^2 + n + 2 \to \Theta(n^2)$

````python
def errepikatua_ote(z,x):
    k = 0
    for y in z :
        if x==y :
            k += 1
            if k == 2:
                return True
    return False
def errepikatu_kopurua(z):
    z2 = []
    for x in z :
        if errepikatua_ote(z,x) and x not in z2 :
            z2.append(x)
    return len(z2)
````

```python
def errepikatua_ote(z,x):
    k = 0                     # 1
    for y in z :              # 2x1  <-->  nx1
        if x==y :             # 2x1  <-->  nx1
            k += 1            # 2x1  <-->  1
            if k == 2:        # 2x1  <-->  1
                return True   # 1    <-->  0
    return False              # 0    <-->  1 
```
