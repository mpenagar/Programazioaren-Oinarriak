Zenbaki zerrenda batetako elementu bakoitien batura

````python
def bakoitien_batura(z):
    b = 0
    for x in z :
        if x%2 == 1 :
            b += x
    return b
````

Problemaren tamaina: $n=len(z)$

```python
def bakoitien_batura(z):
    b = 0              # 1
    for x in z :       # n x 1
        if x%2 == 1 :  # | n x 1
            b += x     # |
    return b           # 1
```

$t(n) = 2n + 2$ &rarr; orden zehatz lineala

````python
def bakoitien_batura(z):
    return sum(x for x in z if x%2 == 1 )
````

Problemaren tamaina: $n=len(z)$

* `sum(it)` &rarr; it-ren elementu kopuru pausu + 1
* `(x for x in ...)` generadorea:
   * sortzea &rarr; 1  
   * zeharkatzea &rarr; `next(g)` 
      * elementu guztiak bakoitiak &rarr; 1 pausu x n aldiz
      * elementu guztiak bikoitiak &rarr; n pausu x 1 aldiz
   
```python
def bakoitien_batura(z):

```

* elementu guztiak bakoitiak &rarr; $t(n) = 2n + 1$
* elementu guztiak bikoitiak &rarr; $t(n) = n + 2$
