```python
class IlaraZirkularra(object):
        
    def __init__(self, size, it=()):
        self.z = [None]*size
        self.len = 0
        self.front = 0
        self.back = size-1
        for x in it :
            self.enqueue(x)

    def __values(self):
        for i in range(self.front,self.front+self.len):
            yield self.z[i % len(self.z)]
            
    def __values(self):
        return (self.z[i % len(self.z)] for i in range(self.front,self.front+self.len))    
            
    def __str__(self):
        return str(list(self.__values()))
    
    def __repr__(self):
        return f'IlaraZirkularra({list(self.__values())})'
    
    def __eq__(self,other):
        # KONTUZ!!!!
        #return type(other) == IlaraZirkularra and self.z == other.z
        return type(other) == IlaraZirkularra and all( x==y for x,y in zip(self.__values(),other.__values()))
    
    def __len__(self):
        return self.len
    
    def isFull(self):
        return self.len == len(self.z)
    
    def enqueue(self,value):
        if self.isFull() :
            raise ValueError("Ilara betea dago")
        self.len += 1
        self.back = (self.back + 1) % len(self.z)
        self.z[self.back] = value

    def dequeue(self):
        if len(self) == 0 :
            raise ValueError("Ilara hutsik dago")
        value = self.frontElement()
        self.len -= 1
        self.front = (self.front + 1) % len(self.z)
        return value
    
    def __iter__(self):
        while self :
            yield self.dequeue()  

    def frontElement(self):
        return self.z[self.front]
    
iz = IlaraZirkularra(20,"kaixo")
print(iz)
for x in " ea hau":
    iz.enqueue(x)
print(iz)
for i in range(9):
    print('dequeue:',iz.dequeue())
print(iz)
for x in range(15):
    iz.enqueue(x)
print(iz)
print(*iz)
print(len(iz))
print(len(iz))
```
