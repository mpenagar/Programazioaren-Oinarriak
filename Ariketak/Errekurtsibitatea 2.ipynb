## 7 
Karaktere kate baten barnean beste kate baten agerpen kopurua (gainezarmenik gabe) kalkulatuko duen funtzioa.

```python
def agerpenak(s1,s2):
    if len(s2) < len(s1) :
        return 0
    if s2[:len(s1)] == s1 :
        return 1 + agerpenak(s1,s2[len(s1):])
    else :
        return agerpenak(s1,s2[1:])
```

```python
agerpenak("froga","xxfrogaxxfrogaxx")
```

```python
agerpenak("kaka","xxxkakakaxxxkakaxx")
```

```python
def agerpenak_aux(s1,s2,i):
    if len(s2)-i < len(s1) :
        return 0
    if s2[i:i+len(s1)] == s1 :
        return 1 + agerpenak_aux(s1,s2,i+len(s1))
    else :
        return agerpenak_aux(s1,s2,i+1)

def agerpenak(s1,s2):
    return agerpenak_aux(s1,s2,0)
```

```python
agerpenak("froga","xxfrogaxxfrogaxx")
```

```python
agerpenak("kaka","xxxkakakaxxxkakaxx")
```

### 8

$a^n$ espresioa kalkulatuko duen funtzioa, a (zeroren ezberdina den) zenbaki erreal bat izanda eta n zenbaki oso bat.

```python
def berreketa(a,n):
    if n == 0 :
        # kasu basea
        return 1.0
    elif n < 0 :
        # kasu orokorra 1
        return 1 / berreketa(a,-n)
    else :
        # kasu orokorra 2
        return a * berreketa(a,n-1)
```

```python
# ongi dabilela frogatze
berreketa(3.54,5)-3.54**5,berreketa(3.54,-5)-3.54**-5
```

Sortu `hurrengoa(i,j,k,n)` funtzio bat zeinak hurrengo konbinaketa indizeak (3-kotea) bueltatuko dituen, edo `None` besterik ez badago.

```python
def hurrengoa(i,j,k,n):
    if k+1 < n :
        return i,j,k+1
    elif j+2 < n :
        return i,j+1,j+2
    elif i+3 < n :
        return i+1,i+2,i+3
    return None
    
```

```python
def konbinaketak(z):
    indizeak = (0,1,2)
    while indizeak :
        i,j,k = indizeak
        print(z[i],z[j],z[k])
        indizeak = hurrengoa(i,j,k,len(z))

def konbinaketak_aux(z,i,j,k):
    print(z[i],z[j],z[k])
    indizeak = hurrengoa(i,j,k,len(z))
    if indizeak :
        konbinaketak_aux(z,i,j,k)
def konbinaketak(z):
    konbinaketak_aux(z,0,1,2)
```
