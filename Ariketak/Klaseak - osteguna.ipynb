```python
class MySet(object):
    """Multzo bat adierazten duen klasea (multzo batetan ezin
    da elementu errepikaturik gorde)."""
    
    def __init__(self,it=()):
        """Multzo bat hasieratuko du iteragarri baten elementu guztiekin"""
        self.z = []
        self.addall(it)
                
    def add(self,value):
        """elementu berri bat gehitzen du multzo batetara,
        aurrez elementu hori ez bazegoen."""
        if value not in self.z :
            self.z.append(value)
        
    def addall(self,it):
        """iteragarri batetako elementu guztiak gehitzen dizkio"""
        for x in it :
            self.add(x)
            
    def __len__(self): 
        """multzoko elementu kopurua bueltatzen du"""
        return len(self.z)
    
    def __str__(self):
        return '{' + ', '.join(repr(x) for x in self.z) + '}'
    
    def __repr__(self):
        return f'MySet({repr(self.z)})'
    
    def __contains__(self,value):
        """multzoan balio bat ote dagoen konprobatzen du"""
        return value in self.z
    
    def __iter__(self): 
        """multzoko elementuen gaineko iteradore bat bueltatuko
        duen funtzioa"""
        return iter(self.z)
    
    def __add__(self,other):
        """bi multzoen bilketa den multzo berria bueltatzen du."""
        m = MySet(self)
        m.addall(other)
        return m
    
    def __eq__(self,other):
        """multzoa beste multzo baten berdina ote den bueltatzen du."""
        return len(self)==len(other) and all(x in self for x in other)
    
    def __ne__(self,other):
        """multzoa beste multzo baten ezberdina ote den bueltatzen du."""
        return len(self)!=len(other) or any(x not in self for x in other)
    
    def __sub__(self,other): 
        """multzoari beste multzoko elementu guztiak kentzean
        lortzen den multzo berria bueltatzen du."""
        m = MySet(self)
        m.removeall(other)
        return m
    
    def remove(self,value):
        """elementu bat ezabatzen du. Elementua ez badago, ez du errorerik sortzen"""
        try :
            self.z.remove(value)
        except ValueError :
            pass
    
    def removeall(self,it): 
        """iteragarri batetako elementu guztiak ezabatzen ditu"""
        for x in it:
            self.remove(x)
```

```python
m = MySet([1,2,3,1,2,3,1,2,3,1,2,3,'kaixo'])

print(f'str(m) : {str(m)}')
print(f'repr(m) : {repr(m)}')
print(f'm : {m}')
print(f'type(eval(repr(m))) : {type(eval(repr(m)))}')
print(f'3 in m : {3 in m}')
print(f'4 in m : {4 in m}')
print(f'list(m) : {list(m)}')
print(f'm + MySet(range(3,7)) : {m + MySet(range(3,7))}')
print(f'm == eval(repr(m)) : {m == eval(repr(m))}')
print(f'm == MySet([3,2,1,"kaixo"]) : {m == MySet([3,2,1,"kaixo"])}')
print(f'm != MySet(range(1,3)) : {m != MySet(range(1,3))}')
print(f'm.remove(1) : {m.remove(1)}')
print(f'm : {m}')
print(f'm.remove(10) : {m.remove(10)}')
print(f'm : {m}')
print(f'm.removeall(range(10)) : {m.removeall(range(10))}')
print(f'm : {m}')
print(f'MySet(range(10)) - MySet(range(0,20,2)) : {MySet(range(10)) - MySet(range(0,20,2))}')
```
