```python
def rrotc(c,n):
    alfa = "abcdefghijklmnÃ±opqrstuvwxyz"
    #i = alfa.find(c)
    i = alfa.index(c)
    j = (i+n)%len(alfa)
    return alfa[j]
    
def rrot(s,n):
    e = ""
    for c in s :
        e = e + rrotc(c,n)
    return e
    
def lrot(s,n):
    return rrot(s,-n)


lrot("asier",2)
```
