Demagun `s` letra xehez osoturiko karaktere kate bat daukazula, bi funtzio idatzi `lrot(s,n)` eta `rrot(s,n)`. Funtzio hauek `s` kateko hizkiak alfabetoraren barnean `n` aldiz ezkerrera (_lrot_) edo eskumara (_rrot_) errotatzen direnean emandako karaktere kateak bueltatuko dituzte.  

```python
def rrotc(c,n):
    alfa = "abcdefghijklmn単opqrstuvwxyz"
    #i = alfa.find(c)
    i = alfa.index(c)
    j = (i+n)%len(alfa)
    return alfa[j]
    
def rrot(s,n):
    e = ""
    for c in s :
        e = e + rrotc(c,n)
    return e
    
def lrot(s,n):
    return rrot(s,-n)


lrot("asier",2)
```

```python
def sorted_string(s):
    for i in range(len(s)-1):
        if s[i] > s[i+1] :
            return False
    return True

def sorted_string(s):
    for a,b in zip(s,s[1:]):
        if a > b :
            return False
    return True

def sorted_string(s):
    #z = list(s)
    #z.sort()
    #z = sorted(s)
    #return z == list(s)
    return sorted(s) == list(s)
    

def sorted_string_alphabet(s):
    alpha = "abcdefghijklmn単opqrstuvwxyz"
    for a,b in zip(s,s[1:]):
        if alpha.index(a) > alpha.index(b) :
            return False
    return True


print(sorted_string("aeiou"))
print(sorted_string("aeiiiou"))
print(sorted_string("aeiaou"))
print(sorted_string("n単o"))
print("-"*40)
print(sorted_string_alphabet("aeiou"))
print(sorted_string_alphabet("aeiiiou"))
print(sorted_string_alphabet("aeiaou"))
print(sorted_string_alphabet("n単o"))
```
