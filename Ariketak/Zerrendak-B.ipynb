```python
def matrize_berria(n, m, x=0):
    A = []
    for i in range(n):
        z = []
        for _ in range(m):
            z.append(x)
        A.append(z)
    return A

def matrize_berria(n, m, x=0):
    A = []
    for _ in range(n):
        z = [x] * m
        A.append(z)
    return A

m = matrize_berria(2,4,-1)
m[1][3] = 1000
m
```
