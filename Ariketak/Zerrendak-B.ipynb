### a - Zenbaki konstante batez (defektuz 0) betetako nxm tamainako matrizea sortu

```python
def matrize_berria(n, m, x=0):
    A = []
    for i in range(n):
        z = []
        for _ in range(m):
            z.append(x)
        A.append(z)
    return A

def matrize_berria(n, m, x=0):
    A = []
    for _ in range(n):
        z = [x] * m
        A.append(z)
    return A

# KONTUZ HONEKIN!!!!!
#def matrize_berria(n, m, x=0):
#    return [[x] * m] * n

m = matrize_berria(2,4,-1)
print(m)
m[1][3] = 1000
print(m)
```

### d - Matrize bat pantailatik idatzi

```python
def printm(A):
    for i in range(len(A)):
        for j in range(len(A[0])):
            print(A[i][j], end='\t')
        #print("\n", end='')
        print()
            
m = matrize_berria(2,4,3)
m = [[1000,2,3],[4,5,6]]
printm(m)
```
