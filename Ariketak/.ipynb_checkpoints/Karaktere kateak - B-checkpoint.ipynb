Demagun `s` letra xehez osoturiko karaktere kate bat daukazula, bi funtzio idatzi `lrot(s,n)` eta `rrot(s,n)`. Funtzio hauek `s` kateko hizkiak alfabetoraren barnean `n` aldiz ezkerrera (_lrot_) edo eskumara (_rrot_) errotatzen direnean emandako karaktere kateak bueltatuko dituzte.  

```python
def rrotc(c,n):
    alfa = "abcdefghijklmn√±opqrstuvwxyz"
    i = alfa.index(c)
    #i = alfa.find(c)
    j = (i + n) % len(alfa)
    return alfa[j]

def rrot(s,n):
    e = ""
    for c in s :
        e = e + rrotc(c,n)
    return e

def rrot(s,n):
    z = []
    for c in s :
        z.append(rrotc(c,n))
    return "".join(z)

print(rrot("abedcf",2)) 

def lrotc(c,n):
    return rrotc(c,-n)

def lrot(s,n):
    return rrot(s,-n)

print(lrot("abedcf",2)) 

rrot(lrot("sdlafikuhglfghsdflkghsldfhg",2364347),2364347)
```
