```python
class IlaraZirkularra(object):
        
    def __init__(self, size, it=()):
        self.__z = [None]*size
        self.__len = 0
        self.__front = 0
        self.__back = size-1
        for x in it :
            self.enqueue(x)

    def __values(self):
        for i in range(self.__front,self.__front+self.__len) :
            yield self.__z[i % len(self.__z)]

    def __values(self):
        return (self.__z[i % len(self.__z)] for i in range(self.__front,self.__front+self.__len))            
            
    def __str__(self):
        return str(list(self.__values()))
    
    def __repr__(self):
        return f'IlaraZirkularra({list(self.__values())})'
    
    def __eq__(self,other):
        # KONTUZ!!!!
        #return type(other) == IlaraZirkularra and self.__z == other.__z        
        return type(other) == IlaraZirkularra and all( x==y for x,y in zip(self.__values(),other.__values()))
    
    def __len__(self):
        return self.__len
    
    def isFull(self):
        return self.__len == len(self.__z)
    
    def enqueue(self,value):
        if self.isFull() :
            raise ValueError("Is Full")
        self.__len += 1
        self.__back = (self.__back + 1) % len(self.__z)
        self.__z[self.__back] = value
        
    def dequeue(self):
        if not self :
            raise ValueError("Is Empty")
        self.__len -= 1
        value = self.__z[self.__front]
        self.__front = (self.__front + 1) % len(self.__z)
        return value
    
    def __iter__(self):
        while self :
            yield self.dequeue()  

    def front(self):
        return self.__z[self.__front]
    
iz = IlaraZirkularra(20,"kaixo")
print(iz)
for x in " ea hau":
    iz.enqueue(x)
print(iz)
for i in range(9):
    print('dequeue:',iz.dequeue())
print(iz)
for x in range(15):
    iz.enqueue(x)
print(iz)
print(*iz)
print(len(iz))
print("----------------------")
# Hau ez litzateke posible izan behar, baina eremu "pribatuak" ikus daitekeenez, 
# ilararen barne egitura ikus dezakegu
print(iz._IlaraZirkularra__z)
print(iz._IlaraZirkularra__len,iz._IlaraZirkularra__front,iz._IlaraZirkularra__back)
```
