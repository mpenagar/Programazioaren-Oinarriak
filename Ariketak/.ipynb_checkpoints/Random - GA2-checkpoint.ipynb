### 6
random moduloko random funtzioa soilik erabiliz, sortu ezazu $[a,b)$ tartean uniformeki 
banatutako zenbaki erreala bueltatuko duen `uniform(a,b)` funtzioa

```python
def uniform(a,b):
    r = random()
    return (b-a)*r+a
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
n = 1000
xa = 20
xb = 30
ya = -35
yb = 40
x = [uniform(xa,xb) for i in range(n)]
y = [uniform(ya,yb) for i in range(n)]
fig = plt.plot(x, y, '.')
```

### 1 
random moduloko random funtzioa soilik erabiliz, sortu ezazu $[a,b]$ tartean uniformeki 
banatutako zenbaki osoa bueltatuko duen `randint(a,b)` funtzioa

```python
import math
def randint(a,b):
    r = uniform(a,b+1)
    return math.floor(r)
def randint(a,b):
    r = uniform(a-1,b)
    return math.ceil(r)
def randint(a,b):
    r = uniform(a-0.5,b+0.5)
    return round(r)
# int funtzioa asimetrikoa da x>0 eta x<0 denean
#def randint(a,b):
#    r = uniform(a,b)
#    return int(r)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
n = 1000
xa = 10
xb = 15
ya = -8
yb = 4
x = [randint(xa,xb) for i in range(n)]
y = [randint(ya,yb) for i in range(n)]
fig = plt.plot(x, y, '.')
```
