Date klasea. Date motako objektu batek data bat errepresentatuko du, urte, hilabete, egun, ordu, minutu eta segunduz osotua. Ondoko funtzioak sortu:

* `__init__(self,year,month,day,hour,minute,second)`
* `__str__(self)` eta `__repr__(self)`
* `offset(self,year,month,day,hour,minute,second)` : data bat desplazatuko du emandako kantitateetan.
* `next_day(self)` : data bat egun batetan aurrera desplazatuko du.
* `previous_day(self)` : data bat egun batetan atzera desplazatuko du.
* `__sub__(self,other)`: bi daten artean pasa diren segundu kopurua bueltatuko du.
* `__add__(self,other)`: data bati segundu kopuru bat batzerakoan lortzen den data berria bueltatuko du.

**OHARRA1**: Data bat inplementatzeko modurik errexenetarikoa, data ezagun batetik pasa diren segunduak soilik erabiltzea da, adibidez 1. urteko urtarrilaren 1-eko 00:00:00-etatik pasatako segundu kopurua.

**OHARRA2**: Suposatu urte bisurterik ez dagoela eta hilabete guztiek 30 egun dituztela

```python
class Date(object):
    def __init__(self,year,month,day,hour,minute,second):
        # erreferentzia: 1/1/1 00:00:00
        self.s = 0
        self.offset(year-1,month-1,day-1,hour,minute,second)

    def offset(self,year,month,day,hour,minute,second):
        self.s += second 
        self.s += minute * 60
        self.s += hour * 60 * 60
        self.s += day * 24 * 60 * 60
        self.s += month * 30 * 24 * 60 * 60
        self.s += year * 12 * 30 * 24 * 60 * 60
        
    def next_day(self):
        #self.offset(0,0,1,0,0,0)
        self.s += 24 * 60 * 60
        
    def previous_day(self):
        #self.offset(0,0,-1,0,0,0)
        self.s -= 24 * 60 * 60
        
    def __sub__(self,other):
        return self.s - other.s
    
    def __add__(self,other):
        d = Date(1,1,1,0,0,0)
        d.s += self.s
        d.s += other
        return d
    
    def to_values(self):
        k = self.s
        Y = 1 + k // (12 * 30 * 24 * 60 * 60)
        k = k % (12 * 30 * 24 * 60 * 60)
        M = 1 + k // (30 * 24 * 60 * 60)
        k = k % (30 * 24 * 60 * 60)
        D = 1 + k // (24 * 60 * 60)
        k = k % (24 * 60 * 60)
        h = 0 + k // (60 * 60)
        k = k % (60 * 60)
        m = 0 + k // 60
        s = k % 60        
        return Y,M,D,h,m,s
    
    def __str__(self):
        Y,M,D,h,m,s = self.to_values()
        return f'{Y}/{M}/{D} {h}:{m}:{s} [{self.s}]'

    def __repr__(self):
        Y,M,D,h,m,s = self.to_values()
        return f'Date({Y},{M},{D},{h},{m},{s})'
    
    def __eq__(self, other):
        return type(other) == Date and self.s == other.s 

d = Date(2026,5,7,9,4,23)
print(d)
#for _ in range(25):
#    d.next_day()
#    print(d)

print(repr(d))
print(eval(repr(d)))

print(eval(repr(d)) == d)
print(d == 3)
```
