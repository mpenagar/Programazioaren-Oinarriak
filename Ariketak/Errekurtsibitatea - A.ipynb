Txikitik handira ordenatutako zenbaki zerrenda batetan zenbaki baten posizioa (edo None, ez badago) bueltatuko duen funtzioa, bilaketa dikotomikoa deritzon teknika erabiliz:
 Tarte batetan bilatzeko, tarteko erdiko elementua begiratuko dugu.
 Elementu hau bilatzen ari garen balioa bada, topatu dugu.
 Ez bada, bilatzen ari garen balioa handiagoa bada eskuinean bilatzen jarraituko dugu eta bestela ezkerrean.

```python
# KASU BASEA AMAITU GABE DAGO,
# ELEMENTUA ZERRENDAN EZ DAGOENEKO KASUETARAKO
def bilatu(z,x):
    e = len(z)//2
    # Kasu basea: elementua erdian dago
    if z[e] == x :
        return e
    else : # kasu generala: ez dago erdian
        if x < z[e] : #ezkerrean bilatu z[:e] zerrendan          
            return bilatu(z[:e],x)
        else : # eskuman bilatu z[e+1:] zerrendan
            i = bilatu(z[e+1:],x)
            if i == None :
                return None
            else :
                return e+1+ i 
            #return None if i==None else e+1+i 
```

```python
z = [1,3,4,6,8,12,34,46,68,987]
all([z[bilatu(z,x)]==x for x in z])
```

```python
z = [1,3,4,6,8,12,34,46,68,987]
bilatu(z,55)
```
