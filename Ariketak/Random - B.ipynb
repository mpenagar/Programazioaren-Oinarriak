```python
def randrange(start, stop=None, step=1):
    # 1. zerrenda bat sortu tarteko elementu guztiekin
    if stop == None:
        start,stop = 0,start
    if start >= stop :
        raise ValueError('start >= stop')
    z = []
    i = start
    while i < stop :
        z.append(i)
        i += step
    # 2. Auzazko elementu bat aukeratu
    pos = randint(0,len(z)-1)
    return z[pos]
    
    
randrange(7,33)
    
```

```python
def randrange(start, stop=None, step=1):
    # 1. zerrenda bat sortu tarteko elementu guztiekin
    if stop == None:
        start,stop = 0,start
    if start >= stop :
        raise ValueError('start >= stop')
    z = list(range(start,stop,step))
    # 2. Auzazko elementu bat aukeratu
    pos = randint(0,len(z)-1)
    return z[pos]
    
    
randrange(7,33)
    
```
