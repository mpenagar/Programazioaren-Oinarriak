## 7
Hiztegi edo multzorik erabili gabe, zerrenda batetan agertzen diren elementu ezberdinen kopurua.

```python
def elementu_ezberdin_kopurua(z):
    aux = []
    for x in z :
        if x not in aux :
            aux.append(x)
    return len(aux)
```

Problemaren tamaina: $n=len(z)$

* `x not in aux`:
    
* `aux.append(x)` &rarr; 1 pausu
* `len(aux)` &rarr; 1 pausu

$t(n)=n+2$

## 8
Hiztegi edo multzoak erabiliz, zerrenda batetan agertzen diren elementu ezberdinen kopurua.

```python
def elementu_ezberdin_kopurua(z):
    return len(set(z))
```

Problemaren tamaina: $n=len(z)$

* `set(z)` &rarr; $n$ pausu
* `len(set)` &rarr; 1 pausu

$t(n)=n+2$
