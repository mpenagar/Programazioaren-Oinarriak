## 7
Hiztegi edo multzorik erabili gabe, zerrenda batetan agertzen diren elementu ezberdinen kopurua.

````python
def elementu_ezberdin_kopurua(z):
    aux = []
    for x in z :
        if x not in aux :
            aux.append(x)
    return len(aux)
````

Problemaren tamaina: $n=len(z)$

* `x not in aux`:
    * kasu ona: z-ko elementu guztiak berdinak &rarr; 1 pausu
    * kasu txarra: z-ko elementu guztiak ezberdinak &rarr; 1, 2, 3,... pausu
* `aux.append(x)` &rarr; 1 pausu
* `len(aux)` &rarr; 1 pausu

**Kasu ona**: z-ko elementu guztiak berdinak 

```python
def elementu_ezberdin_kopurua(z):
    aux = []                # 1
    for x in z :            # n
        if x not in aux :   # n
            aux.append(x)   # 1
    return len(aux)         # 1
```

$t(n) \approx n+1$

**Kasu txarra**: z-ko elementu guztiak ezberdinak

```python
def elementu_ezberdin_kopurua(z):
    aux = []
    for x in z :
        if x not in aux :
            aux.append(x)
    return len(aux)
```

$t(n) \approx n^2+n+1$
