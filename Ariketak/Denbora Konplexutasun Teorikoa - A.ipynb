## 7
Hiztegi edo multzorik erabili gabe, zerrenda batetan agertzen diren elementu ezberdinen kopurua.

````python
def elementu_ezberdin_kopurua(z):
    aux = []
    for x in z :
        if x not in aux :
            aux.append(x)
    return len(aux)
````

Problemaren tamaina: $n=len(z)$

* `x not in aux`:
    * kasu ona: z-ko elementu guztiak berdinak &rarr; 1 pausu
    * kasu txarra: z-ko elementu guztiak ezberdinak &rarr; 1, 2, 3,... pausu
* `aux.append(x)` &rarr; 1 pausu
* `len(aux)` &rarr; 1 pausu

**Kasu ona**: z-ko elementu guztiak berdinak 

```python
def elementu_ezberdin_kopurua(z):
    aux = []                # 1
    for x in z :            # n
        if x not in aux :   # n
            aux.append(x)   # 1
    return len(aux)         # 1
```

$t(n) \approx n+1$

**Kasu txarra**: z-ko elementu guztiak ezberdinak

```python
def elementu_ezberdin_kopurua(z):
    aux = []
    for x in z :
        if x not in aux :
            aux.append(x)
    return len(aux)
```

$t(n) \approx n^2+n+1$

## 9
Hiztegi edo multzorik erabili gabe, zerrenda batetan errepikatuak (bi edo gehiagotan) agertzen diren elementu ezberdinen kopurua.

Erabiliko dugun barne egitura:

```python
hz = [ [g1,b1] , [g2,b2] , ...]
```
non g zerrendako balioak diren eta b ber agerpen kopurua

````python
def errepikatu_ezberdin_kopurua(z):
    hz = []
    for x in z :
        for gb in hz :
            if gb[0] == x :
                gb[1] += 1
                break
        else :
            hz.append([x,1])
    k = 0
    for g,b in hz :
        if b > 1 :
            k += 1
    return k
    retutn sum(1 for g,b in hz if b > 1)
    
````

## 11
Hiztegi edo multzorik erabili gabe, testu fitxategi batetan hutsunez banandutako hitz ezberdinen kopurua.

````python
def hitz_ezberdin_kopurua(bideizena, kodifikazioa="utf8"):
    with open(bideizena, encoding=kodifikazioa) as f:
        z = []                         # 1
        for line in f :                #  | n x 1
            for word in line.split():  #  | 
                if word not in z :     #  |[nx1,1+2+3+..+n]
                    z.append(word)     #  |
        return len(z)                  #  1
````

Problemaren tamaina: $n=hitz\ kopurua$

* **Kasu ona**: hitz guztiak berdinak 
   * $t(n) = 2n + 2$ &rarr; $\Omega(n)$
* **Kasu txarra**: hitz guztiak ezberdinak 
   * $t(n) = \frac{n(n+1)}{2} + n + 2$ &rarr; $O(n^2)$
   

## 12
Hiztegi edo multzoak erabiliz, testu fitxategi batetan hutsunez banandutako hitz ezberdinen kopurua

````python
def hitz_ezberdin_kopurua(bideizena, kodifikazioa="utf8"):
    with open(bideizena, encoding=kodifikazioa) as f:
        s = set()                      # 1
        for line in f :                #  | n x 1
            for word in line.split():  #  | 
                s.add(word)            #  | n x 1
        return len(s)                  #  1
````

Problemaren tamaina: $n=hitz\ kopurua$

* $t(n) = 2n + 2$ &rarr; $\Theta(n)$
