```python
from random import randrange 
def sample(population, k):
    if k > len(population):
        raise ValueError(f'{k} > len(population) = {len(population)}')
    z = []
    while len(z) < k :
        i = randrange(len(population))
        #print(i)
        if i not in z :
            z.append(i)
    z2 = []
    for i in z :
        z2.append(population[i])
    return z2
        
sample("abcdefghi", 4)
#sample("aaaaaabbbbbbbcdef", 4)
```

```python
from random import suffle 
def sample(population, k):
    if k > len(population):
        raise ValueError(f'{k} > len(population) = {len(population)}')
    z = list(range(len(population)))
    suffle(z)
    for i in z[:k] :
        z2.append(population[i])
    return z2
        
```
