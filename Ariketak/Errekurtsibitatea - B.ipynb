Karaktere kate baten barnean beste kate baten agerpen kopurua (gainezarmenik gabe) kalkulatuko duen funtzioa.


```python
def agerpen_kopurua(x,s):
    #print(x,s)
    if len(x) > len(s) : # kasu basea
        return 0
    else : # kasu orokorra
        #if s[:len(x)] == x :
        if s.startswith(x) :
            return 1 + agerpen_kopurua(x,s[len(x):])
        else :
            return agerpen_kopurua(x,s[1:])
```

```python
agerpen_kopurua("aa","aaeiouaaaaeaaa")
```

```python
def agerpen_kopurua(x,s):
    return _agerpen_kopurua(x,s,0)


def _agerpen_kopurua(x,s,i):
    #print(x,s)
    if i+len(x) > len(s) : # kasu basea
        return 0
    else : # kasu orokorra
        if s[i:i+len(x)] == x :
            return 1 + _agerpen_kopurua(x,s,i+len(x))
        else :
            return _agerpen_kopurua(x,s,i+1)
        
agerpen_kopurua("aa","aaeiouaaaaeaaa")
```
