```python
def agerpen_kopurua(x,s):
    #print(x,s)
    if len(x) > len(s) : # kasu basea
        return 0
    else : # kasu orokorra
        #if s[:len(x)] == x :
        if s.startswith(x) :
            return 1 + agerpen_kopurua(x,s[len(x):])
        else :
            return agerpen_kopurua(x,s[1:])
```
