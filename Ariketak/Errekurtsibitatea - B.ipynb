Karaktere kate baten barnean beste kate baten agerpen kopurua (gainezarmenik gabe) kalkulatuko duen funtzioa.


```python
def agerpen_kopurua(x,s):
    if x == s : # kasu basea
        return 1
    else : # kasu orokorra
        #if s[:len(x)] == x :
        if s.startswith(x) :
            return 1 + agerpen_kopurua(x,s[len(x):])
        else :
            return agerpen_kopurua(x,s[1:])
```
