## Funtzio generadoreak
* Funtzio arruntak
* Funtzioa exekutatu &rarr; generadore berri bat
* `return`erabili beharrean, `yield` sententzia

```python
def arruntak10():
    for i in range(1,11):
        print(i)
```

```python
arruntak10()
```

```python
def arruntak10():
    for i in range(1,11):
        yield i
```

```python
arruntak10()
```

```python
g = arruntak10()
print(*g)
print('Berriro:',*g)
```

```python
print(*arruntak10())
print('Berriro:',*arruntak10())
```

```python
def arruntak(n):
    for i in range(1,n+1):
        print(i)
```

```python
def arruntak(n):
    for i in range(1,n+1):
        yield i
```

```python
g = arruntak(17)
print(*g)
print(*g)
print(*arruntak(17))
```

```python
def arruntak():
    i = 1
    while True :
        print(i)
        i += 1
```

```python
#arruntak()
```

```python
def arruntak():
    i = 1
    while True :
        yield i
        i += 1
```

```python
g = arruntak()
print(*zip(g,"abcdefghijk"))
```

Zenbaki lehenen (guztien) sekuentzia:

```python
def lehenak():
    i = 2
    yield i
    z = [i]
    while True :
        i += 1
        # zerrendako zenbakiren batengatik zatigarria?
        #if not any(i%j==0 for j in z):
        if all(i%j!=0 for j in z):
            yield i
            z.append(i)
```

```python
for _,i in zip(range(10),lehenak()):
    print(i)
    
```

Iteragarri batetako lehenengo elementuak:

```python
def lehenengoak(it,n=10):
    for _,x in zip(range(n),it):
        yield x
```

```python
print(*lehenengoak(lehenak(),15))
```
