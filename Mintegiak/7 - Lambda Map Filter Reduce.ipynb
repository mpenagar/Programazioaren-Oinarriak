# map 

```python
map(func, *iterables)
```

````python
help(map)
````

````python
it = range(10)
f = lambda x : 2*x
map(f,it)
````

````python
list(map(f,it))
````

````python
print(*map(f,it))
````

````python
list(map(lambda x : 2*x, range(10)))
````

````python
print(*map(lambda x : 2*x, range(10)))
````

````python
[ 2*x for x in range(10) ]
````

````python
( 2*x for x in range(10) )
````

````python
g = ( 2*x for x in range(10) )
list(g)
````

````python
g = ( 2*x for x in range(10) )
print(*g)
````

````python
m = map(lambda i,s: i*s, range(1,6), "aeiou")
list(m)
````

````python
def my_zip(it1,it2):
    return map((lambda a,b : (a,b)), it1, it2)
````

````python
def bildu(a,b):
    return (a,b)

def my_zip(it1,it2):
    return map(bildu, it1, it2)
````

````python
my_zip = lambda it1,it2: map((lambda a,b : (a,b)), it1, it2)
````

# Filter 

```python
filter(func, iterable)
```

````python
help(filter)
````

````python
bikoiti = lambda x : x%2==0
it = range(10)
filter(bikoiti,it)
````

````python
f = filter(bikoiti,it)
list(f)
````

````python
g = ( x for x in it if bikoiti(x) )
list(g)
````

### erredukzio funtzioak

`sum` , `max` , `min`, `any` , `all`

````python
it = range(1,11)
````

````python
batuketa = lambda x,y : x+y
print(sum(it))
reduce(batuketa,it)
````

````python
batuketa = lambda x,y : x+y
batukaria = lambda it : reduce(batuketa,it,0)
batukaria(it)
````

````python
def batuketa(x,y):
    return x+y
def batukaria(it):
    return reduce(batuketa,it,0)
batukaria(it)
````

````python
#batukaria(("kjgf","kjyhfk","ljhgk"))
````

````python
biderkaketa = lambda x,y : x*y
biderkaria = lambda it : reduce(biderkaketa,it,1)
biderkaria(it)
````

````python
faktoriala = lambda n : biderkaria(range(1,n+1))
faktoriala(10)
````

````python
handia = lambda x,y : x if x>y else y
maximoa = lambda it : reduce(handia,it)
maximoa([23,41,241,24,235,13,51,35,13,51,2,152,4132,412,4,451,25,23])
````

````python
bitakobat = lambda x,y : x or y
batenbat = lambda it : reduce(bitakobat,it)
batenbat([False,True,False,False])
````

````python
batenbat([False,False,False,False])
````

````python
biak = lambda x,y : x and y
denak = lambda it : reduce(biak,it)
denak([False,True,False,False])
````

````python
denak([True,True,True,True])
````
