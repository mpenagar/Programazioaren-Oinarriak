#### 1 - Multzoen `add(e)`-en baliokidea

````python
class myset():
    
    def __init__(self, init_size=10):
        self.t = [[] for _ in range(init_size)]
        
    def add(self,value):
        z = self.t[hash(value) % len(self.t)]
        for x in z :
            if x == value :
                break
        else :
            z.append(value)
````

````python
s = myset()
for e in ("Ane","Jon","Miren","Asier","Jon","Jon","Jon"):
    s.add(e)
print(s.t)
````

#### 3 - Multzoen `len()`-en baliokidea

````python
class myset():
    
    def __init__(self, iterable=tuple(), init_size=10): # t(n) = 1 + "len(iterable)"
        self.t = [[] for _ in range(init_size)]
        for x in iterable :
            self.add(x)
   
    def add(self,value): # t(n) = 1
        z = self.t[hash(value) % len(self.t)]
        for x in z :
            if x == value :
                break
        else :
            z.append(value)
        
    def __len__(self): # t(n) = n/4
        return sum(len(z) for z in self.t)
````

#### 4 - ITERAGARRITASUNA: Multzoen `iter()`-en baliokidea

````python
class myset():
    
    def __init__(self, iterable=tuple(), init_size=10): # t(n) = 1 + "len(iterable)"
        self.t = [[] for _ in range(init_size)]
        self.n = 0
        for x in iterable :
            self.add(x)
   
    def add(self,value): # t(n) = 1
        z = self.t[hash(value) % len(self.t)]
        for x in z :
            if x == value :
                break
        else :
            z.append(value)
            self.n += 1
        
    def __len__(self): # t(n) = 1
        return self.n
    
    def __iter__(self): # t(n) = 1
        for z in self.t :
            for x in z :
                #print(x)
                yield x
````

````python
s = myset(["Ane","Jon","Miren","Asier","Nora","Aiala","Eneritz"])
print(s.t)
print(*s)
print(list(s))
````
